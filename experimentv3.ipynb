```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import requests
import itertools
from __future__ import division
from operator import itemgetter
from pandas.io.json import json_normalize
from scipy import special, optimize
```

```python
# openFDA API key: TTwz56GNTpZnsL4il0pLIwz7F43QObZOQEssWlSU
drugname1 = 'tylenol'
drugname2 = 'ibuprofen'
drugname3 = 'aspirin'
# top n adverse effects
n = 8

def get_drug_adverse_event_data(drugname):
    
    api_key = 'TTwz56GNTpZnsL4il0pLIwz7F43QObZOQEssWlSU&search'
    request_string='https://api.fda.gov/drug/event.json?api_key=' + api_key
    request_string= request_string + '=patient.drug.medicinalproduct:' + drugname
    request_string= request_string +'&count=patient.reaction.reactionmeddrapt.exact'
    
    json_df = requests.get(request_string).json()
    json_adverse_list = json_df['results']
    
    total100 = 0
    for i in json_adverse_list:
        total100 += i['count']
    for i in json_adverse_list:
        i['count'] = float(i['count']/total100)
    
    return json_adverse_list
    
def get_event_freq(event_list, event):
    try:
        index=map(itemgetter('term'), event_list).index(event)
        return event_list[index].get('count')
    except ValueError:
        return 0
```

```python
dlist1=get_drug_adverse_event_data(drugname1)
dlist2=get_drug_adverse_event_data(drugname2)
dlist3=get_drug_adverse_event_data(drugname3)

reacts_topn  = [None] * n
d1_freq_list = [None] * n
d2_freq_list = [None] * n
d3_freq_list = [None] * n

for i in range(n):
    reacts_topn[i] = dlist1[i]['term']
reacts_topn[3] = dlist1[17]['term'] # correction for plot
for i in range(n):
    d1_freq_list[i]=get_event_freq(dlist1, reacts_topn[i])
    d2_freq_list[i]=get_event_freq(dlist2, reacts_topn[i])
    d3_freq_list[i]=get_event_freq(dlist3, reacts_topn[i])
```

```python
# Conditional probability of effect given drug and adverse events
width = 0.2
x = np.array([1,2,3,4,5,6,7,8])
my_xticks = reacts_topn
plt.subplots()
plt.xticks(x, my_xticks)
rects1 = plt.bar(x, d1_freq_list, width, color='r')
rects2 = plt.bar(x+width, d2_freq_list, width, color='y')
rects3 = plt.bar(x+(2*width), d3_freq_list, width, color='g')
plt.legend((rects1[0], rects2[0], rects3[0]),(drugname1, drugname2, drugname3))
plt.setp(plt.xticks()[1], rotation=45)
plt.show()
```

```python
count1=[]
count2=[]
count3=[]

for i in range(n):
    count1.append(get_event_freq(dlist1, reacts_topn[i]))
    count2.append(get_event_freq(dlist2, reacts_topn[i]))
    count3.append(get_event_freq(dlist3, reacts_topn[i]))
    
ind=np.arange(len(count1))
                  
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(25,25))

axes[0].barh(ind, count1, color='r')
axes[0].set_yticks(ind+.3)
axes[0].set_yticklabels(reacts_topn, fontsize=20)
axes[0].set_title(drugname1, fontsize=30)

axes[1].barh(ind, count2, color='y')
axes[1].set_yticks(ind+.3)
axes[1].set_yticklabels(reacts_topn, fontsize=20)
axes[1].set_title(drugname2, fontsize=30)

axes[2].barh(ind, count3, color='g')
axes[2].set_yticks(ind+.3)
axes[2].set_yticklabels(reacts_topn, fontsize=20)
axes[2].set_title(drugname3, fontsize=30)

plt.suptitle('Probability of effect given drug', fontsize=30)

plt.xticks(fontsize=20)
plt.show()
#plt.savefig('AAAfdaOpen_incubator_fig_02.png')
```

```python
'''
For many drugs and many events we can use a 3D bar plot based on 
http://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html#bar-plots
Matplotlib has a 3d plotting extension 'mplot3d', which supports many type of 3D plots
'''
from mpl_toolkits.mplot3d import Axes3D

colors=['b','g','r', 'c', 'm', 'y']
drugs=['paxil', 'lexapro', 'hydrocodone', 'xanax', 'tramadol', 'vicodin']
events=['FLUSHING', 'DYSPNOEA', 'FATIGUE', 'NAUSEA', 'PAIN', 'DIZZINESS', 'HEADACHE']
positions=[50,40,30,20,10,0]
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
for c, drug, z in zip(colors, drugs, positions):
    xs = np.arange(len(events))
    ys = []
    druglist=get_drug_adverse_event_data(drug)
    
    for event in events:
        ys.append(get_event_count(druglist, event))

    ax.bar(xs, ys, zs=z, zdir='y', color=c, alpha=0.8)

ax.set_xticklabels(events)
ax.set_yticklabels(drugs)
ax.set_xlabel('Events')
ax.set_ylabel('Drugs')
ax.set_zlabel('Event Count')

plt.title('Adverse Events', fontsize=20)
plt.show()
```
